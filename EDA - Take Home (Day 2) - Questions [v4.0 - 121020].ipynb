### About the dataset (Types of Flower Data)

The dataset contains the information of the three different species of the iris flower. The dataset contains values of four features (sepal length, sepal width, petal length, and petal width) of each flower and the species of iris.

Attribute Information:

**sepal length:** Sepal length of a flower in cm

**sepal width:** Sepal width of a flower in cm

**petal length:** Petal length of a flower in cm

**petal width:** Petal width of a flower in cm

**class:** Species of the iris flower

```python
# type your code here
df_flower1=df_flower.select_dtypes(include=np.number)
correlation_matrix=df_flower1.corr()
rows,columns=correlation_matrix.shape

correlation_matrix_boolean=pd.DataFrame(columns=correlation_matrix.columns,index=correlation_matrix.index)
for i in range(rows):
    for j in range(columns):
        if correlation_matrix.iloc[i,j]>0.8:
            correlation_matrix_boolean.iloc[i,j]=True
        else:
            correlation_matrix_boolean.iloc[i,j]=False

print(correlation_matrix_boolean)
# rows,columns=correlation_matrix.shape
```

```python
# type your code here
plt.figure(figsize=(8, 6))
sns.ecdfplot(data=df_flower, x='sepal length', color='b')
plt.axvline(x=6, color='r', linestyle='--', label='Sepal Length = 6 cm')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('ECDF (Probability)')
plt.title('ECDF of Sepal Length')
plt.legend()
plt.grid(True)

# Calculate the probability that sepal length is less than 6 cm
probability_less_than_6 = df_flower['sepal length'][df_flower['sepal length'] < 6].count() / len(df_flower)
print("Probability that Sepal Length is less than 6 cm:", probability_less_than_6)

plt.show()
```
